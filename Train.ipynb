# Test

```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
model_path = '/content/gdrive/MyDrive/Colab Notebooks/image-segmentation-yolov8/runs/segment/train/weights/last.pt'
img_path = '/content/gdrive/MyDrive/Colab Notebooks/image-segmentation-yolov8/tes/hello.jpeg'
```

```python
img = cv2.imread(img_path)
```

```python
H, W, _ = img.shape
model = YOLO(model_path)
results = model(img)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
```

```python
for result in results:
    if result.masks is not None:
        for j, mask in enumerate(result.masks.data):
            mask = mask.numpy() * 255
            mask = cv2.resize(mask, (W, H))

            plt.imshow(mask, cmap='gray')
            plt.title(f'Segmentation Mask {j + 1}')

plt.show()
```
